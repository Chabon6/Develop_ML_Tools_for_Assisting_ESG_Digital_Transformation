```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris


df = pd.read_csv("TEST_ML.csv")
```

```python
def kmeans_(df):
    from sklearn.cluster import KMeans
    
    columns_to_include = ['enginesizel', 'cylinders', 'fueltype']

    X = df[columns_to_include]  # x為特徵資料
    
    global model
    model = KMeans(n_clusters = 39, n_init='auto', random_state=1)  # 預計分為5群，迭代次數由模型自行定義
    model.fit(X)  # 建立模型

    df['Cluster'] = model.labels_  #將分類結果加回df
    
    return df
```

```python
def classic(auto_class):
    # 在 DataFrame 中新增一個欄位來儲存群集標籤
    df['cluster'] = model.labels_

    # 選擇要顯示的群集標籤（將 0 替換為感興趣的群集）
    cluster_label_to_display = auto_class

    # 根據指定的群集標籤篩選 DataFrame，獲取屬於該群集的資料點
    cluster_data = df[df['cluster'] == cluster_label_to_display]

    # 顯示指定群集中的資料點
    return cluster_data
```

```python
'''觀察資料，得出不同群集可能為什麼分類'''
class_base = classic(0)
class_mid = classic(2)
class_top = classic(1)
```
